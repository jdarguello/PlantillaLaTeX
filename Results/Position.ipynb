```python
text = ''
with open('pos1.txt', 'r') as file:
    text = file.readlines()
    for i in range(len(text)):
        text[i] = text[i].replace('\n', '')
        fila = []
        word = ''
        for palabra in text[i]:
            if palabra != ',':
                word += palabra
            else:
                try:
                    fila.append(float(word))
                except:
                    fila.append(word)
                word = ''
        try:
            fila.append(float(word))
        except:
            fila.append(word)
        text[i] = fila
text[0]
```

```python
dic = {}
for i in range(4):
    if text[i][0][0] == 'T':
        dic[text[i][0][:-1]] = []
    else:
        for j in range(1,len(text[i])):
            dic[text[i][0][0] + str(j)] = []   
for i in range(len(text)):
    for j in range(1,len(text[i])):
            if text[i][0][0] == 'T':
                dic[text[i][0][:-1]].append(text[i][j])
            else:
                dic[text[i][0][0] + str(j)].append(text[i][j])
```

```python
with open('posicion1.txt', 'w+') as file:
    suma = 0
    maximo = 0
    for key in dic:
        if len(dic[key]) > maximo:
            maximo = len(dic[key])
    for key in dic:
        if len(dic[key]) == maximo:
            file.write(key)
            suma += 1
            if suma < len(dic):
                file.write('\t')
    file.write('\n')
    for i in range(0,len(dic['x1']),10):
        for key in dic:
            if len(dic[key]) == maximo:
                file.write(str(dic[key][i]))
                if i < len(dic['x1']):
                    file.write('\t')
                else:
                    file.write('\n')
```
